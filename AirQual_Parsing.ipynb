```python
df6 = pd.read_csv('Health_RespiratoryHosp.csv')

# Load data
df6 = pd.read_csv('Health_RespiratoryHosp.csv')

# Drop unnecessary columns
df6 = df6.drop(columns=["GeoTypeDesc", "GeoID", "GeoRank", "Estimated annual number"])

# Rename column
df6 = df6.rename(columns={"Estimated annual rate per 100,000 adults": "Respiratory Hospitalizations per 100000"})

# Convert TimePeriod to string
df6['TimePeriod'] = df6['TimePeriod'].astype(str)

# Function to expand rows with year ranges
def expand_years(row):
    time_period = row['TimePeriod']
    if '-' in time_period:  # If it's a range (e.g., 2017-2019)
        start_year, end_year = map(int, time_period.split('-'))
        return [{**row.to_dict(), 'TimePeriod': str(year)} for year in range(start_year, end_year + 1)]
    else:  # If it's a single year
        return [row.to_dict()]

# Apply function to all rows and flatten the list
expanded_rows = [new_row for _, row in df6.iterrows() for new_row in expand_years(row)]

# Create new DataFrame
df6 = pd.DataFrame(expanded_rows)

# Keep only rows where TimePeriod is in the range 2019-2022
df6 = df6[df6['TimePeriod'].astype(int).between(2019, 2022)]

# Reset index
df6 = df6.reset_index(drop=True)

df6
```

```python
import numpy as np

def calculate_aqhi_updated(row):
    NO2 = row["Nitric Dioxide (NO2) - ppb"]
    O3 = row["Ozone (O3) - ppb"]
    PM2_5 = row["PM2.5 - mcg/m3"]
    
    aqhi = (10 / 10.4) * 100 * (
        (np.exp(0.000537 * O3) - 1) +
        (np.exp(0.000871 * NO2) - 1) +
        (np.exp(0.000487 * PM2_5) - 1)
    )
    return aqhi

# Apply the updated function to each row
df["AQHI"] = df.apply(calculate_aqhi_updated, axis=1)

# Select relevant columns for the output DataFrame
df_aqhi_updated = df[["TimePeriod", "Geography", "AQHI"]]
aqhi = df_aqhi_updated
aqhi
```

```python
import numpy as np
aqi_breakpoints = {
    "PM2.5": [(0.0, 9.0, 0, 50), (9.1, 35.4, 51, 100), (35.5, 55.4, 101, 150),
               (55.5, 125.4, 151, 200), (125.5, 225.4, 201, 300), (225.5, 500.4, 301, 500)],
    "NO2": [(0, 53, 0, 50), (54, 100, 51, 100), (101, 150, 101, 150), 
            (151, 200, 151, 200), (201, 300, 201, 300)],
    "O3": [(0.000, 0.054, 0, 50), (0.055, 0.070, 51, 100), (0.071, 0.085, 101, 150),
           (0.086, 0.105, 151, 200), (0.106, 0.200, 201, 300)]
}

# Function to compute AQI for a given pollutant
def compute_aqi(value, breakpoints):
    for (low, high, ilow, ihigh) in breakpoints:
        if low <= value <= high:
            return round(((ihigh - ilow) / (high - low)) * (value - low) + ilow)
    return None

# Calculate AQI for each pollutant
df["PM2.5 AQI"] = df["PM2.5 - mcg/m3"].apply(lambda x: compute_aqi(x, aqi_breakpoints["PM2.5"]))
df["NO2 AQI"] = df["Nitric Dioxide (NO2) - ppb"].apply(lambda x: compute_aqi(x, aqi_breakpoints["NO2"]))
df["O3 AQI"] = df["Ozone (O3) - ppb"].apply(lambda x: compute_aqi(x / 1000, aqi_breakpoints["O3"]))  # Convert ppb to ppm
df["NO AQI"] = df["Nitric Oxide (NO) - ppb"].apply(lambda x: compute_aqi(x, aqi_breakpoints["NO2"]))  # Using NO2 breakpoints

# Final AQI is the max AQI from all pollutants
df["AQI"] = df[["PM2.5 AQI", "NO2 AQI", "O3 AQI", "NO AQI"]].max(axis=1)

# Group AQI by year and region
aqi_summary = df.groupby(["TimePeriod", "Geography"])["AQI"].max().reset_index()

aqi_summary.to_csv()
aqi_summary
```

```python
import matplotlib.pyplot as plt
import numpy as np
aqhi_summary = aqhi
# Convert to numeric (ensure the data is valid)
df5['Adult Asthma Hospitalizations per 10000'] = pd.to_numeric(df5['Adult Asthma Hospitalizations per 10000'], errors='coerce')
aqhi_summary['AQHI'] = pd.to_numeric(aqhi_summary['AQHI'], errors='coerce')

# Clean data by dropping NaNs (important for the trendline fitting)
df5 = df5.dropna(subset=['Adult Asthma Hospitalizations per 10000'])
aqhi_summary = aqhi_summary.dropna(subset=['AQHI'])

# Merge datasets (use 'inner' to only keep rows that exist in both)
df_merged = df5.merge(aqhi_summary, on=['TimePeriod', 'Geography'], how='inner')

# Check the merged dataframe
print(f'Merged DataFrame shape: {df_merged.shape}')
print(df_merged.head())

# Get the unique years from 'TimePeriod'
years = df_merged['TimePeriod'].unique()

# Plot a scatter plot for each year with a trendline
plt.figure(figsize=(12, 10))
for year in years:
    # Filter data by the current year
    df_year = df_merged[df_merged['TimePeriod'] == year]
    
    # Scatter plot of AQI vs Asthma Hospitalizations for this year
    plt.scatter(df_year['AQHI'], df_year['Adult Asthma Hospitalizations per 10000'], alpha=0.7, label=f"Year {year} Data")
    
    # Apply polyfit for trendline (linear regression)
    z = np.polyfit(df_year['AQHI'], df_year['Adult Asthma Hospitalizations per 10000'], 1)  # Linear fit
    p = np.poly1d(z)
    
    # Sort data for the trendline so the line is plotted in order
    sorted_indices = np.argsort(df_year['AQHI'])  # Sort AQI values
    sorted_aqi = df_year['AQHI'].iloc[sorted_indices]
    sorted_hospitalizations = p(sorted_aqi)  # Calculate the trendline values
    
    # Plot the trendline for this year
    plt.plot(sorted_aqi, sorted_hospitalizations, label=f"Trend Line {year}", linestyle='--')

# Set labels, title, and grid
plt.xlabel('Air Quality Health Index (AQHI)')
plt.ylabel('Asthma Hospitalizations per 10,000')
plt.title('AQHI vs Asthma Hospitalizations for Different Years')
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(rotation=45)

# Set y-axis gridlines and labels to show multiples of 5
y_min = int(df_merged['Adult Asthma Hospitalizations per 10000'].min())
y_max = int(df_merged['Adult Asthma Hospitalizations per 10000'].max())
plt.yticks(np.arange(y_min-1, y_max , 5), fontsize=10)
plt.ylim(bottom=0)

# Show the legend
plt.legend()

# Show the plot
plt.show()
```

```python
# Calculate and display the correlation coefficient for each year

# Dictionary to store correlation coefficients
correlation_results = {}

# Get unique years
years = df_merged['TimePeriod'].unique()

# Initialize plot
plt.figure(figsize=(12, 10))

for year in years:
    # Filter data by the current year
    df_year = df_merged[df_merged['TimePeriod'] == year]
    
    # Scatter plot
    plt.scatter(df_year['AQHI'], df_year['Adult Asthma Hospitalizations per 10000'], alpha=0.7, label=f"Year {year} Data")
    
    # Calculate and store the correlation coefficient
    correlation = np.corrcoef(df_year['AQHI'], df_year['Adult Asthma Hospitalizations per 10000'])[0, 1]
    correlation_results[year] = correlation
    
    # Apply polyfit for trendline
    z = np.polyfit(df_year['AQHI'], df_year['Adult Asthma Hospitalizations per 10000'], 1)
    p = np.poly1d(z)
    
    # Sort data for the trendline
    sorted_indices = np.argsort(df_year['AQHI'])
    sorted_aqi = df_year['AQHI'].iloc[sorted_indices]
    sorted_hospitalizations = p(sorted_aqi)
    
    # Plot the trendline
    plt.plot(sorted_aqi, sorted_hospitalizations, linestyle='--', label=f"Trend Line {year}")

# Display correlation coefficients
for year, corr in correlation_results.items():
    print(f"Year {year}: Correlation Coefficient = {corr:.3f}")

# Set labels, title, and grid
plt.xlabel('Air Quality Health Index (AQHI)')
plt.ylabel('Asthma Hospitalizations per 10,000')
plt.title('AQHI vs Asthma Hospitalizations for Different Years')
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(rotation=45)

# Set y-axis gridlines and labels
plt.yticks(np.arange(int(df_merged['Adult Asthma Hospitalizations per 10000'].min()) - 1,
                     int(df_merged['Adult Asthma Hospitalizations per 10000'].max()), 5), fontsize=10)
plt.ylim(bottom=0)

# Show the legend
plt.legend()

# Show the plot
plt.show()
```
